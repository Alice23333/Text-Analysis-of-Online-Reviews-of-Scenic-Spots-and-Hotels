```python
import pandas as pd
```

# 景区评论

```python
scenic_comment = pd.read_csv("E:/泰迪杯/数据挖掘2021/C/C题数据/景区评论.CSV",encoding='gbk')
```

```python
scenic_comment
```

```python
#无缺失值
#438条重复值（不管日期）
scenic_comment[['景区名称','评论内容']].duplicated().sum()
```

```python
#查看部分重复值
scenic_comment[['景区名称','评论内容']][scenic_comment.duplicated()]
```

```python
#删除重复值
scenic_comment.drop_duplicates(subset={'景区名称','评论内容'},keep="first",inplace=True)
scenic_comment.reset_index(inplace=True)
scenic_comment.drop(labels=['index'],axis=1,inplace=True)
```

```python
scenic_comment
```
