```python
import pandas as pd
```

```python
review_path = 'E:/泰迪杯/数据挖掘2021/C/C题数据/清洗过的数据/景区评论utf-8.csv'
topic_path = 'E:/泰迪杯/数据挖掘2021/LDA结果/景区.csv'
result_path = 'E:/泰迪杯/数据挖掘2021/LDA结果/LDA分主题结果/'
```

```python
reviews = pd.read_csv(review_path, encoding='utf-8')
topic = pd.read_csv(topic_path, encoding='utf-8')[['Dominant_Topic','Topic_Perc_Contrib','Keywords','Text']]
```

```python
# 换列名
name = {'Dominant_Topic':"主题",'Topic_Perc_Contrib':'评论对主题贡献度','Keywords':'主题关键字','Text':'评论分词'}
topic.rename(columns = name,inplace=True)
```

```python
# 主题数据中没有包含最后一条评论，把reviews最后一条删掉
reviews.drop([len(reviews)-1],inplace=True)
```

# 划分主题

```python
result = reviews.join(topic)
```

```python
# 更改主题
result['评价方面'] = '其他'
```

```python
# 服务（1，5，6），位置（10，11），设施（2，8，9，12），卫生（0），性价比（3，4，7）
# 输入一行数据
def change_topic(x):
    if x[6] == '[]':  # 评论分词为空
        return x
    if x[3] == 1.0 or x[3] == 5.0 or x[3] == 6.0:  
        x[7] = '服务'
    elif x[3] == 10.0 or x[3] == 11.0:
        x[7] = '位置'
    elif x[3] == 2.0 or x[3] == 8.0 or x[3] == 9.0 or x[3] == 12.0:
        x[7] = '设施'
    elif x[3] == 0.0:
        x[7] = '卫生'
    elif x[3] == 3.0 or x[3] == 4.0 or x[3] == 7.0:
        x[7] = '性价比'
    return x
```

```python
result = result.apply(lambda x:change_topic(x), axis=1)
```

```python
result
```

```python
# 除去评论分词为空的列表
result = result[result['评价方面'] != '其他']
```

```python
result.to_csv(result_path + '景区（不含名词）utf-8.csv', encoding='utf-8',index=False)
result.to_csv(result_path + '景区（不含名词）gbk.csv', encoding='gbk',index=False)
```
