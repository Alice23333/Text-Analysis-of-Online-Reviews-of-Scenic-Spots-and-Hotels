# 景区评论

```python
scenery = pd.read_csv("E:/泰迪杯/数据挖掘2021/C/C题数据/清洗过的数据/景区评论.csv",encoding='gbk')
```

```python
scenery
```

```python
path = "E:/泰迪杯/数据挖掘2021/C/C题数据/景区评论unicode/"
for i in range(1,10):
    filter = 'A0'+ str(i)
    tmp = scenery[scenery['景区名称']==filter]
    path1 = path + "A0" + str(i) + ".csv"
    tmp.to_csv(path1, encoding='utf-8', index=False)
```

```python
path = "E:/泰迪杯/数据挖掘2021/C/C题数据/景区评论unicode/"
for i in range(10,51):
    filter = 'A'+ str(i)
    tmp = scenery[scenery['景区名称']==filter]
    path1 = path + "A" + str(i) + ".csv"
    tmp.to_csv(path1, encoding='utf-8', index=False)
```
